there are two types of columns here, categorical and numeric all non numeric 'object' type columns are categorical ones. first we deal with missing values of numeric columns

handle nan and missing values 

```python
# Handling missing values for numerical columns by replacing with mean
numerical_cols = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed', 
                  'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 
                  'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']
```

```python
for col in numerical_cols:
    if data[col].dtype != 'object':  # Check if column is numerical
        mean_value = data[col].mean()
        data[col].fillna(mean_value, inplace=True)
```

handling categorical columns

one hot encoding is to convert categorical columns into binary

```python
# Assuming 'data' is your DataFrame and 'categorical_cols' contains the categorical column names
categorical_cols = ['WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']

# Replace missing categorical values with the mode (most frequent value) of each column
for col in categorical_cols:
    mode_val = data[col].mode()[0]  # Calculate mode value
    data[col].fillna(mode_val, inplace=True)  # Replace missing values with mode
```

```python

# Perform one-hot encoding for categorical columns
data = pd.get_dummies(data, columns=categorical_cols)

# # Convert all columns to numeric
# for col in data.columns:
#     data[col] = pd.to_numeric(data[col], errors='coerce')

# # Now all columns will be converted to numeric types where possible
```

```python


# Define a function to split the data
def train_test_split_custom(data, test_size=0.2):
    # Calculate the index to split the data
    split_index = int(len(data) * (1 - test_size))
    
    # Shuffle the indices for random selection
    indices = np.random.permutation(data.index)
    
    # Split the indices into train and test sets
    train_indices, test_indices = indices[:split_index], indices[split_index:]
    
    # Create X_train, Y_train, X_test, Y_test
    X_train, X_test = data.iloc[train_indices].drop(columns=['RainTomorrow']), data.iloc[test_indices].drop(columns=['RainTomorrow'])
    Y_train, Y_test = data.iloc[train_indices]['RainTomorrow'], data.iloc[test_indices]['RainTomorrow']
    
    return X_train, Y_train, X_test, Y_test


```

```python
# Perform the split
X_train, Y_train, X_test, Y_test = train_test_split_custom(data, test_size=0.2)

# Save the datasets into separate CSV files
X_train.to_csv('X_train.csv', index=False)
Y_train.to_csv('Y_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)

Y_test.to_csv('Y_test.csv', index=False)
```
